## 1. Quick Stats

```python
# Show how many ingredients in each fridge
for cuisine, ingredients in fridges.items():
    print(f"{cuisine:12} fridge has {len(ingredients):2} ingredients")
```

## 2. Find Common Ingredients Across All Cuisines

```python
# Find ingredients that appear in ALL fridges
def find_universal_ingredients():
    # Start with first fridge
    common = set(fridges['Indian'])
    
    # Keep only items that appear in all fridges
    for ingredients in fridges.values():
        common = common.intersection(set(ingredients))
    
    return sorted(list(common))

universal = find_universal_ingredients()
print(f"Ingredients found in EVERY fridge: {len(universal)}")
print(universal)
```

## 3. Most Popular Ingredients Overall

```python
# Count how many times each ingredient appears across all fridges
all_ingredients = []
for ingredients in fridges.values():
    all_ingredients.extend(ingredients)

ingredient_counts = Counter(all_ingredients)

print("Top 20 Most Common Ingredients:")
for ingredient, count in ingredient_counts.most_common(20):
    print(f"{ingredient:30} appears in {count}/9 cuisines")
```

## 4. Unique Ingredients Per Cuisine

Find ingredients that are ONLY in one cuisine's fridge

```python
def find_unique_ingredients(cuisine_name):
    """Find ingredients unique to one cuisine."""
    my_ingredients = set(fridges[cuisine_name])
    
    # Get all ingredients from other cuisines
    other_ingredients = set()
    for name, ingredients in fridges.items():
        if name != cuisine_name:
            other_ingredients.update(ingredients)
    
    # Find what's unique
    unique = my_ingredients - other_ingredients
    return sorted(list(unique))

# Show unique ingredients for each cuisine
for cuisine in fridges.keys():
    unique = find_unique_ingredients(cuisine)
    print(f"\n{cuisine} unique ingredients ({len(unique)}):")
    print(unique[:5])  # Show first 5
```

## 5. Random Recipe Generator

Pick a cuisine and get random ingredients for a recipe!

```python
def generate_random_recipe(cuisine_name, num_ingredients=5):
    """Generate a random recipe from a cuisine."""
    ingredients = fridges[cuisine_name]
    selected = random.sample(ingredients, min(num_ingredients, len(ingredients)))
    
    print(f"\nüç≥ Random {cuisine_name} Recipe Idea:")
    print("="*40)
    for i, ingredient in enumerate(selected, 1):
        print(f"{i}. {ingredient}")
    print("\nNow go create something delicious!")

# Try it out
generate_random_recipe('Indian', 7)
generate_random_recipe('Korean', 6)
generate_random_recipe('Mexican', 5)
```

## 6. Fusion Cuisine Generator

Mix two cuisines together!

```python
def fusion_recipe(cuisine1, cuisine2, num_ingredients=6):
    """Create a fusion recipe from two cuisines."""
    ingredients1 = fridges[cuisine1]
    ingredients2 = fridges[cuisine2]
    
    # Take half from each
    half = num_ingredients // 2
    selected = random.sample(ingredients1, half) + random.sample(ingredients2, num_ingredients - half)
    random.shuffle(selected)
    
    print(f"\nüåç {cuisine1}-{cuisine2} Fusion Recipe:")
    print("="*40)
    for i, ingredient in enumerate(selected, 1):
        print(f"{i}. {ingredient}")

# Try some fusion combos
fusion_recipe('Korean', 'Mexican')
fusion_recipe('Indian', 'French')
fusion_recipe('American', 'Malaysian')
```
